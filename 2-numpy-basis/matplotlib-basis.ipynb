# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x,cosy,color="red")
plt.plot(x,siny)
plt.show()
```

```python
plt.plot(x,cosy,color="red",linestyle="--")
plt.plot(x,siny)
plt.show()
```

```python
plt.plot(x,cosy,color="red",linestyle="--")
plt.plot(x,siny)
plt.xlim(-5,15)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
plt.plot(x,cosy,color="red",linestyle="--")
plt.plot(x,siny)
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)")
plt.plot(x,siny,label="sin(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("ML World")
plt.show()
```

### 散点图

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
plt.show()
```
